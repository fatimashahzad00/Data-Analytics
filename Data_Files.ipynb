```python
from sklearn import datasets
iris = datasets.load_iris()
print(iris.data)
```

```python
print(iris.DESCR)
```

```python
print(iris.data.shape)
print(iris.feature_names)
print(iris.target)
print(iris.target.shape)
print(iris.target_names)
```

```python
print(type(iris.data))
```

```python
import pandas as pd
df = pd.DataFrame(iris.data)
df.head()
```

```python
df.columns = iris.feature_names
df.head()
```

```python
df.describe()
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
for i in df.columns:
    sns.histplot(df[i])
    plt.title(i)
    plt.show()
```

```python
for i in df.columns:
  print(i, "Skewness: ", df[i].skew())
  print('----------------------------')
  print(i, "Kurtosis: ", df[i].kurt())
```

```python
import numpy as np
import pandas as pd
colors = list()
palette = {0: 'red', 1: 'green', 2: 'blue'}
```

```python
for c in np.nditer(iris.target):
    colors.append(palette[int(c)])

dataframe = pd.DataFrame(iris.data, columns = iris.feature_names)
sc = pd.plotting.scatter_matrix(dataframe, alpha=0.3, figsize=(10,10), diagonal='hist', color=colors, marker = 'o', grid = True)

```
